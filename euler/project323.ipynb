```python
import numpy as np
from scipy.special import binom
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def calcul(n):
    s=1
    p=1
    q=2
    for j in range(1,n+1):
        s+=binom(n,j)*p/(q-1)
        p*=-1
        q*=2
    return s
```

```python
plt.close()
L= [calcul(n) for n in range(10,70)]
plt.plot(range(10,70),L)
plt.show()
```
