```python
import numpy as np
from copy import deepcopy
from random import choice
```

```python
def sous_carre(i,j):
    return ((i-i%3)//3,(j-j%3)//3)

def prerempli(M):
    Z=[[[k for k in range(1,10)] for _ in range(9)] for _ in range(9)]
    for i in range(9):
        for j in range(9):
            if M[i,j]!=0:
                Z[i][j]=[int(M[i,j])]
            else:
                
                for k in range(9):
                    try:
                        Z[i][j].remove(M[i,k])
                    except:
                        pass
                    try:
                        Z[i][j].remove(M[k,j])
                    except:
                        pass
                    a,b = sous_carre(i,j)
                    try:
                        Z[i][j].remove(M[3*a+k%3,3*b+(k-k%3)//3])
                    
                    except:
                        pass
    return Z


def actualise(Z,i,j,val): # on a donné la valeur a en i j sur la matrice Z
    for k in range(9):
        try:
            Z[i][k].remove(val)
        except:
            pass
        try:
            Z[k][j].remove(val)
        except:
            pass
        a,b = sous_carre(i,j)
        try:
            Z[3*a+k%3][3*b+(k-k%3)//3].remove(val)
        except:
            pass
        
def test_possible(Z,i,j,val):
    possible = True
    a,b = sous_carre(i,j)
    for k in range(9):
        possible = (Z[i][k] != [val]) and Z[k][j]!=[val] and Z[3*a+k%3][3*b+(k-k%3)//3]!=[val] and possible
    
    return possible

        
        

def est_valide(grille):
    s=prerempli(grille)
    for i in range(9):
        for j in range(9):
            if s[i][j]==[]:

                return False
    print(grille)
    return True



def aux(M,Z,i,j,liste): # renvoie la valeur que prend la li
    if liste == []: # ie que nous n'avons plus rien à voir
        if est_valide(M):
            return True,M[i,j]
        else:
            return False, 0
    print(liste)
    for val in Z[i][j]:
        Q=deepcopy(Z)
        M[i,j]=val
        Q[i][j]=[val]
        actualise(Q,i,j,val)
        a,b= liste.pop()
        possible,valeur = aux(M,Q,a,b,liste)
        
        if possible :
            M[i,j]=valeur
            return True,valeur
    # si on sort de la boucle sans retour c'est qu'aucune valeur n'a été trouvée en l'état de la configuration
    return False,0

# autre heuristique, implémenter si une grille est possible ou non
def solve_sudoku(M):
    base = prerempli(M)
    liste =[] # liste des indices encore en lice
    for i in range(9):
        for j in range(9):
            if len(base)>=2:
                liste.append((i,j))
    indice = liste.pop()
    aux(M,base,i,j,liste)
    
    return M
```
